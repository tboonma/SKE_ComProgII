# Using *slicing* or *fancy indexing* techniques, create a 2d array with 1 on the border and 0 inside

```python
import numpy as np
arr = np.ones((5,5))
for i in range(1, int(len(arr))-1):
    for j in range(1, int(len(arr[i]))-1):
        arr[i][j] = 0
arr
```

# Create a checkerboard (with values 1 and 0 inside) 10x10 matrix using the `tile` function

```python
import numpy as np
arr = np.array([1,0])
arr2 = np.array([0,1])
chk = np.tile([arr, arr2], (5,5))
chk
```

# Write a code to plot both sine and cosine curves in the same plot with proper legend.

Use the code below as your starting point

```python
import matplotlib.pyplot as plt
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)
# Plot the points using matplotlib
plt.plot(x,y, x,z)

plt.xlim(0, 3 * np.pi)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sinusoidal curve(s)')
plt.legend(['Sine','Cosine'])
```

# Subplots

You can display different plots in the same figure using the subplots function. Here is an example:

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, axes = plt.subplots(nrows=2)
axes[0].plot(x, y_sin)
axes[0].set_xlim(0, 3 * np.pi)
axes[0].set_title('Sine')
axes[1].plot(x, y_cos)
axes[1].set_xlim(0, 3 * np.pi)
axes[1].set_title('Cosine')
fig.show()
```

# Plotting

Produce a plot of the functions $f(x)=e^{-x/10}\sin(\pi x)$ and $g(x)=xe^{−x/3}$ over the interval $[0, 10]$. Include labels for the x- and y-axes, and a legend explaining which line is which plot.

Hint:
*   `np.pi`
*   `np.sin()`
*   `np.exp()`



```python
import numpy as np
import matplotlib.pyplot as plt
interval = np.arange(0,10.1,0.1)
f = []
g = []
for x in interval:
    f.append(np.exp(-x/10)*np.sin(np.pi*x))
    g.append(x*np.exp(-x/3))
plt.plot(interval,f, interval,g)
plt.xlim(0,10.1,0.1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Functions')
plt.legend(['f(x)','g(x)'])
```
