# Alcohol Consumption (5 points)



```python
import pandas as pd
```

Import dataset from `alcohol.csv`

```python
alc = pd.read_csv('alcohol.csv')
alc
```

Which continent drinks more beer on average? List average beer consumption per continent.

```python
more_beer = alc[['continent', 'beer_servings']].groupby("continent").mean()
more_beer = more_beer.reset_index()
continent_more_beer = more_beer.sort_values('beer_servings', ascending=False)
continent_more_beer
```

For each continent, `describe()` the statistics for wine consumption.

```python
wine = alc[['continent', 'wine_servings']].groupby("continent").mean()
wine.describe()
```

Show the mean alcohol consumption per continent for every alcohol type using the DataFrame layout below.

image.png

```python
mean_alc = alc.groupby("continent").mean()
mean_alc
```

Show the mean, min and max values for spirit consumption using the DataFrame layout below.

image.png

```python
mean_detail = alc[['continent', 'spirit_servings']].groupby("continent").mean()
min_detail = alc[['continent', 'spirit_servings']].groupby("continent").min()
max_detail = alc[['continent', 'spirit_servings']].groupby("continent").max()
detail = pd.concat([mean_detail, min_detail, max_detail], axis=1)
detail = detail.set_axis(['mean', 'min', 'max'], axis=1)
detail
```

Import titanic passenger data

Note: please upload `titanic2.csv` file using File Explorer tool on your left

**This is different from `titanic.csv` from the previous lab.**

image.png

```python
titanic = pd.read_csv('titanic2.csv')
titanic
```

Set `PassengerId` as the index

Hint: use `set_index()` method

```python
new_titanic = titanic.set_index('PassengerId')
new_titanic
```

Visualize a pie chart representing the gender proportion

example: image.png

```python
import matplotlib.pyplot as plt
proportion = titanic[['PassengerId', 'Sex']].groupby('Sex').count()
proportion = proportion.reset_index()
info = proportion.values.T

labels = info[0]
sizes = info[1]
explode = (0, 0.15)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=325)
ax1.set(title='Sex Proportion')
plt.show()
```

Visualize a scatter plot with the Fare and the Age, each dot colored by gender

example: image.png

```python
import matplotlib.pyplot as plt
data = titanic[['Fare', 'Age', 'Sex']]
data.sort_values('Sex')
gender = list(set(data['Sex'].tolist()))
fig, ax = plt.subplots()
for g in gender:
  x = []
  y = []
  for d in data.values:
    if d[2] == g:
      x.append(d[1])
      y.append(d[0])
  label = g
  if g == 'male':
    color = 'blue'
  elif g == 'female':
    color = 'green'
  else:
    color = 'white'
  ax.scatter(x, y, c=color, label=label)

ax.legend()
plt.title('Fare x Age')
plt.xlabel('Age')
plt.ylabel('Fare')
ax.grid(True)
```

Visualize a histogram with the Fare

example: image.png

```python
data = titanic[['Fare']]
ax = data.plot.hist(bins=100, grid=True)
ax.set_xlim(0, 600)
ax.set_ylim(0, 350)
plt.xlabel('Fare')
plt.title('Fare Payed Histogram')
```

Visualize a bar plot representing survival rates in percent grouped by the passenger class

```python
data = titanic[['Survived', 'Pclass']]
data = data.groupby('Pclass').sum()
data = data.reset_index()
ax = data.plot.bar('Pclass', 'Survived')
plt.ylabel('Survived')
plt.title('survival rates in percent grouped by the passenger class')
```

# Apple Stock (10 points)

Import dataset from `apple_stock_1980_2014.csv`

```python
apple = pd.read_csv('apple_stock_1980_2014.csv')
apple
```

Print the type of each data column

```python
apple.dtypes
```

Convert the Date column to a datetime type

```python
apple['Date'] = pd.to_datetime(apple['Date'])
apple
```

Set the date as the index

```python
apple.set_index('Date')
```

Check if there is any duplicate dates

```python
apple[apple.duplicated(['Date'])]
```

Currently, the index is from the most recent date. Sort the entries so that the oldest date comes first.

```python
new_apple = apple.set_index('Date')
new_apple = new_apple.sort_index()
new_apple
```

Find the last business day of each month using `resample('BM')`

```python
new_apple.resample('BM').last()
```

Compute the difference in days between the first and the last day

```python
from datetime import datetime
new_apple = apple.set_index('Date')
new_apple = new_apple.sort_index()
new_apple = new_apple.reset_index()
last = new_apple.tail(1).loc[:, 'Date'].to_string()
start = new_apple.head(1).loc[:, 'Date'].to_string()
start = start.split('   ')[1]
last = last.split('   ')[1]
start_time = datetime.strptime(start, '%Y-%m-%d')
last_time = datetime.strptime(last, '%Y-%m-%d')
(last_time - start_time).days
```

Compute the number of months in the data we have

```python
number_of_month = (last_time.year - start_time.year)*12 + (last_time.month - start_time.month)
number_of_month
```

Plot the 'Adj Close' value.

```python
data = apple[['Date', 'Adj Close']]
data = data.set_index('Date')
data.plot()
plt.ylabel('Adj Close')
plt.title('Adj close x year')
```
